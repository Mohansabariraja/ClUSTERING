```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
d = pd.read_csv("/content/Mall_Customers.csv")
```

```python
d.head()
```

```python
encoder = LabelEncoder()
d['Genre'] = encoder.fit_transform(d['Genre'])
d.head()
```

```python
x = d.drop(['CustomerID','Genre','Age'], axis=1)
```

```python
standard_scalar = preprocessing.StandardScaler()
x = standard_scalar.fit_transform(x)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 5,linkage = 'average')
d['cluster'] = model.fit_predict(x)
d.head()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(1, figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('dendogram')
plt.xlabel('customers')
plt.ylabel('euclidean')
plt.show()
```

```python
plt.scatter(d['Annual_Income_(k$)'],d['Spending_Score'],c = d['cluster'])
```
